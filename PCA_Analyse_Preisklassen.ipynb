```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import pylab as plt
import numpy as np
from matplotlib import cm
```

```python
df_lu = pd.read_csv("Luxembourg_new.csv")
df_de = pd.read_csv("Deutschland_new.csv")
df_be = pd.read_csv("Belgien_new.csv")
df_fr = pd.read_csv("Frankreich_new.csv")
```

```python
features_lu = ['Distanz','fläche_m2','schlafzimmer','badezimmer','parkplätze','Haus']
features_fr = ['baujahr','Bevölkerung','fläche_m2','parkplätze','badezimmer','Haus']
features_be = ['fläche_m2','badezimmer','parkplätze','Haus']
features_de = ['baujahr','Bevölkerung','DistanzKöln','DistanzLuxemburg','fläche_m2','schlafzimmer','badezimmer']
```

```python
#Daten skalieren

#Features extrahieren
x = df_lu.loc[:, features_lu].values

#Zielvariable extrahieren
y = df_lu.loc[:,['kaufpreis']].values 

# Features Standardizieren
x = StandardScaler().fit_transform(x)
```

```python
#Hauptkomponentenanalyse mit 2 Hauptkomponenten (2D Projektion)
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['PC1', 'PC2'])
finalDf = pd.concat([principalDf, df_lu[['kaufpreis']]], axis = 1)
```

```python
#2D Projektion visualisieren
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Hauptkomponent 1', fontsize = 15)
ax.set_ylabel('Hauptkomponent 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)

targets = [200000,500000,800000,1000000,15000000,2000000]
upperLimits = [500000,800000,1000000,1500000,2000000,40000000]
titles = [">200000",">500000",">800000",">1000000",">15000000",">2000000"]

xyz = np.arange(10)
ys = [i+xyz+(i*xyz)**2 for i in range(10)]
colors = cm.rainbow(np.linspace(0, 1, len(ys)))
pos = 0

for target,limit,color in zip(targets,upperLimits,colors):
    indicesToKeep = (finalDf['kaufpreis'] >= target) & (finalDf['kaufpreis'] <= limit)
    ax.scatter(finalDf.loc[indicesToKeep, 'PC1']
               , finalDf.loc[indicesToKeep, 'PC2']
               , color = color
               , s = 50)
    pos = pos + 1
ax.legend(titles)
ax.grid()

pca.explained_variance_ratio_
```
